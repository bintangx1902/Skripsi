```python
from matplotlib import pyplot as plt

from shared.local_path import *
from shared.utils import preprocess_audio, librosa
```

```python
# 'E:\Skripsi\data\audio\calm\03-01-02-01-01-01-01_aug1.wav'
aug_audio_path = os.path.join(AUDIO_PATH, 'calm', '03-01-02-01-01-01-01_aug1.wav')
audio_path = os.path.join(AUDIO_PATH, 'calm', '03-01-02-01-01-01-01.wav')
```

```python
import matplotlib.pyplot as plt
import librosa.display

def display(original_path: str, augmented_path: str):
    original = preprocess_audio(original_path)
    augmented = preprocess_audio(augmented_path)

    fig, axs = plt.subplots(1, 2, figsize=(40, 12), sharex=True)

    img1 = librosa.display.specshow(original[:, :, 0], sr=32000, x_axis='time', y_axis='mel', ax=axs[0])
    axs[0].set_title('Mel Spectrogram - Original', fontsize=48)
    axs[0].set_ylim(0, 8000)
    axs[0].tick_params(axis='both', labelsize=24)  # Axis tick label size

    cbar1 = fig.colorbar(img1, ax=axs[0], format='%+2.0f dB')
    cbar1.ax.tick_params(labelsize=24)  # Colorbar tick label size

    img2 = librosa.display.specshow(augmented[:, :, 0], sr=32000, x_axis='time', y_axis='mel', ax=axs[1])
    axs[1].set_title('Mel Spectrogram - Augmented', fontsize=48)
    axs[1].set_ylim(0, 8000)
    axs[1].tick_params(axis='both', labelsize=24)

    cbar2 = fig.colorbar(img2, ax=axs[1], format='%+2.0f dB')
    cbar2.ax.tick_params(labelsize=24)

    plt.tight_layout()
    plt.show()
```

```python
display(audio_path, aug_audio_path)
```
